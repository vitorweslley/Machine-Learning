```python
# Importa√ß√£o das bibliotecas necess√°rias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# 1. Carregar o conjunto de dados Iris
iris = load_iris()
X = iris.data  # Caracter√≠sticas (atributos das flores)
y = iris.target  # Classes (tipos de flores)

# 2. Pr√©-processamento: Normaliza√ß√£o dos Dados
# Normalizamos os dados para que o PCA funcione corretamente
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Aplicar PCA para reduzir a dimensionalidade para 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# 4. Comparar a vari√¢ncia explicada pelos componentes principais
explained_variance = pca.explained_variance_ratio_
print("Vari√¢ncia explicada pelos dois primeiros componentes principais:")
print(explained_variance)

# 5. Visualizar os dados ap√≥s o PCA (reduzidos para 2D)
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k')
plt.title("Visualiza√ß√£o dos Dados Iris com PCA (2D)")
plt.xlabel("Primeiro Componente Principal (PC1)")
plt.ylabel("Segundo Componente Principal (PC2)")
plt.colorbar(label='Classe da Flor')
plt.show()

# 6. Visualizar a vari√¢ncia explicada pelos componentes principais
plt.figure(figsize=(8, 6))
plt.bar(range(1, len(explained_variance) + 1), explained_variance, alpha=0.7, color='blue')
plt.title('Vari√¢ncia Explicada pelos Componentes Principais (PCA)')
plt.xlabel('Componente Principal')
plt.ylabel('Vari√¢ncia Explicada')
plt.show()
```
